# 🚀 prism v2.0 - Flat Grid ARC-AGI-3 Solver

Pure Rust genetic algorithm solver with SIMD-friendly flat Grid layout.

**Engine Specs:**
- Flat Grid: `Vec<i32>` for cache-friendly, SIMD-accessible memory
- Hot paths: ColorSwap/Clear use direct slice iteration
- Binary: 633KB stripped ELF with LTO optimization

**Deployment:**
1. Download prism binary from Kaggle Dataset
2. Execute against ARC-AGI-3 test.json
3. Generate submission.json

```python
import subprocess
import os
import sys

DATASET_PATH = "/kaggle/working/prism-binary-dataset"
os.makedirs(DATASET_PATH, exist_ok=True)

print("📥 1. 正在下載 prism v2.0 引擎...")
subprocess.run([
    "kaggle", "datasets", "download", "-d", "nazcawong/prism-binary-dataset",
    "-p", DATASET_PATH, "--unzip"
], check=True)

prism_BIN = f"{DATASET_PATH}/prism"
os.chmod(prism_BIN, 0o755)

print("🚀 2. 啟動 prism 引擎進行全量壓測...")
# 使用 Popen 來即時印出 Rust 引擎的高速運算日誌
process = subprocess.Popen(
    [prism_BIN, "/kaggle/input/arc-prize-2026-arc-agi-3/test.json", "/kaggle/working/submission.json"],
    stdout=subprocess.PIPE,
    stderr=subprocess.STDOUT,
    text=True,
    bufsize=1
)

# 即時捕捉並顯示輸出
for line in process.stdout:
    print(line, end="")

process.wait()

print("\n🏁 3. 執行結束，檢查產出檔案：")
if os.path.exists("/kaggle/working/submission.json"):
    size_kb = os.path.getsize("/kaggle/working/submission.json") / 1024
    print(f"✅ submission.json 成功生成！(大小: {size_kb:.2f} KB)")
else:
    print("❌ 警告：未找到 submission.json！")
```
